```python
# load sample dataset
from pycaret.datasets import get_data
data = get_data('bank')
print(type(data))
```

```python
from pycaret.classification import *
s = setup(data, target = 'default', session_id = 123)
```

```python
# Сравнение моделей
```

```python
# functional API
best = compare_models()
```

```python
print(best)
```

```python
# Analyze Model
```

```python
# functional API
evaluate_model(best)
```

```python
# functional API
plot_model(best, plot = 'auc')
```

```python
# functional API
plot_model(best, plot = 'confusion_matrix')
plot_model(best, plot = 'class_report')
plot_model(best, plot = 'error')
```

```python
# functional API
predictions = predict_model(best, data=data)
predictions.head()
```

```python
# вероятность обоих классов задается через raw_score
# functional API
predictions = predict_model(best, data=data, raw_score=True)
predictions.head()
```

```python
# сохранение модели
# functional API
save_model(best, 'ridge')
```

```python
# Загрузка модели в среду
# functional API
loaded_model = load_model('ridge')
print(loaded_model)
```
